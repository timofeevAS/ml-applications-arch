## Digitraffic

```python
from pprint import pprint
```

```python
from sources.digitraffic import DigitrafficSource

digitraffic_source = DigitrafficSource([20002])

result = digitraffic_source.fetch()
pprint(result)
```

```python
from transformers.digitraffic import DigitrafficTransformer

digitraffic_transformer = DigitrafficTransformer()

transformed_result = digitraffic_transformer.transform(result)
pprint(transformed_result)
```

```python
from publisher.stdout import StdoutPublisher

stdout_publisher = StdoutPublisher()

stdout_publisher.publish(transformed_result)
```

### Example of `Digitraffic` pipeline

```python
digitraffic_source = DigitrafficSource([20002])
source_data = digitraffic_source.fetch()

digitraffic_transformer = DigitrafficTransformer()
transformed_data = digitraffic_transformer.transform(source_data)

stdout_publisher = StdoutPublisher()
stdout_publisher.publish(data=transformed_data)
```
